```python
cc = pd.read_excel('Country-Code.xlsx')
```

```python
cc
```

```python
df3 = pd.merge(data, cc, how = 'left', on = 'Country Code')
```

```python
df3
```

```python
countries = df3.Country

countries.value_counts()
```

```python
def other_label(x):
    count = x.value_counts()
    labels = count[count < 80].index
    
    series = x.apply(lambda s: 'Others' if s in labels else s)
    
    return series
```

```python
newC = other_label(countries)

newC
```

```python
country_labels = newC.value_counts().index

country_labels
```

```python
country_counts = newC.value_counts().values

country_counts
```

```python
plt.pie(country_counts, labels = country_labels, autopct = '%1.2f%%')
```

# Observation
1. India uses Zomato more than any country in the world, with more than 90% users
2. On Second we have USA
3. On Third United Kingdom
4. Other Countries

```python
rating_df = df3.groupby(['Aggregate rating','Rating text', 'Rating color']).size().reset_index().rename(columns = {
    0: 'Rating count'
})

rating_df
```

```python
#find out which country has given the most 0 ratings

df3[df3['Rating color'] == 'White'].groupby(df3.Country).size().reset_index()
```

# Observation
    - The maximum number of 0 ratings are from Indian Customers

```python
# find out which country is using which currency

df3[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index().rename(columns = {
    0:'Count'
})
```

```python
# Which countries have online delivery option

df3[df3['Has Online delivery'] == 'Yes'].groupby('Country').size().reset_index()
```

```python
df3[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```

# Observations
    - Online Deliveries are available in India and UAE

```python
# Create a Pie chart based on the top 5 cities that used Zomato
city_values = df3.City.value_counts().values
city_labels = df3.City.value_counts().index

plt.pie(city_values[:5], labels=city_labels[:5], autopct='%1.2f%%')
```
